```python
import geopandas as gpd
import numpy as np
import pandas as pd
```

```python
roads=gpd.read_file('EU_data/mmr_2/mmr_transport/nieuwe/networks_meters.shp')
bus_centroid=gpd.read_file('EU_data/mmr_2/mmr_transport/infrastructure/nodes/mmr_t_bus_centroids.shp')
ferry_terminal=gpd.read_file('EU_data/mmr_2/mmr_transport/infrastructure/nodes/mmr_t_ferry_terminals.shp')
metro_station=gpd.read_file('EU_data/mmr_2/mmr_transport/infrastructure/nodes/mmr_t_metro_station.shp')
rail_station=gpd.read_file('EU_data/mmr_2/mmr_transport/infrastructure/nodes/mmr_t_rail_station.shp')
```

```python
bus_centroid['ACCTYPE']=1
ferry_terminal['ACCTYPE']=2
metro_station['ACCTYPE']=3
rail_station['ACCTYPE']=4
```

```python
rail_station=rail_station.rename(columns={"Name": "NAME"})
```

```python
nodes=pd.concat([bus_centroid,ferry_terminal,metro_station,rail_station])
```

```python
nodes.to_file('EU_data/mmr_2/mmr_transport/nieuwe/nodes.shp', driver='ESRI Shapefile')
```
