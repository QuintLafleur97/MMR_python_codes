```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
#%matplotlib inline 

landuse_cols = ["Agriculture", "Forest", "Industry",
             "Nature", "Recreation", "Urban-formal",
             "Urban-informal", "Urban-villages", "Transport", "Airport and Harbour",
             "Water"]
```

```python
er1 = pd.read_csv('enrichment_MMR_5.csv').drop(columns=['Unnamed: 0'])
er2 = pd.read_csv('enrichment_MMR_5.csv').drop(columns=['Unnamed: 0'])

name_map={}
for i in range(len(landuse_cols)):
    name_map[i+1] = landuse_cols[i]
name_map
```
