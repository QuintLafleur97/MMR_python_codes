```python
import pandas as pd
import oapackage
```

```python
#cp
df = pd.read_csv(r'C:\\Users\\Quint\\Documents\\EPA\\Thesis\\ENC_Py3_mra-main\\enrichment_final\\ENC_Py3_mra-main_2\\ENC_Py3_mra-main\\EU_output\\mmr_2\\2. theta_cp_coarse_cal_output_2.csv',index_col='theta_cp')
df['Metric_3']=1-df['Metric_3']
df2=df.values
df2=df2.transpose()
pareto=oapackage.ParetoDoubleLong()

for ii in range(0, df2.shape[1]):
    w=oapackage.doubleVector((df2[0,ii], df2[1,ii], df2[2,ii]))
    pareto.addvalue(w, ii)

pareto.show(verbose=1)
list_pareto=pareto.allindices()
df['Pareto']='No'
list_2=['no']*len(df)
for i in list_pareto:
    list_2[i]='Yes'
df['Pareto']=list_2
df
```
